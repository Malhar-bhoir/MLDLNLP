```python
### file handling and exception handling 
try :
    file = open('example.txt', 'r')
    content = file.read()
    print(content)
except FileNotFoundError:
    print('not found')   
except Exception as ex :
    print(ex)
finally:
    if 'file' in locals() or not file.closed():
        file.close()
        print('fileclose')  

```

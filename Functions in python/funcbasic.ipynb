```python

# temprsture conversion functions
# def celsius_to_fahrenheit  
def celsuis_to_fara(temp,unit):
    if unit == "c":
        return (temp * 9/5) + 32 # c to F
    elif unit =="F":
        return (temp - 32) * 5/9 # F to C
    else:
        return None
    
# Example usage
celsuis_to_fara(100, "c")  # Converts 100 degrees Celsius to Fahrenheit
celsuis_to_fara(212, "F")  # Converts 212 degrees Fahrenheit
```

```python
# paswoord strength checker 
# we will chec kif the pass word has any digit or not using the "char.isdigit()"
def password_strength(password):
    if len(password)< 8:
        return False
    elif not any(char.isdigit() for char in password): # we iterste throuhgh each character in pass "char in pass"
        return False
    elif not any(char.islower()for char in password):
        return False
    elif not any(char.isupper() for char in password):
        return False
    elif not any(char in "!@#$%^&*()_+" for char in password): # lloks for s[recial character sin the pass words ]
        return False
    return True


## caliing the finvtion 
print(password_strength("Passwo"))  # True
    
         
 
```

```python
# calcu;ate cost of items in shoppifg carts
#"totalcost"+= is used to add the cost of each item to the total cost 

def calculate_total_cost(cart):
    total_cost = 0.0
    for item in cart :
        total_cost+= item["price"] * item["quantity"]  # calculate cost for each item while iterating 
    return total_cost
# 3 here we are using items["price "] as we are usinf the dictoniers 

cart = [

    {"item": "apple", "price": 1.0, "quantity": 3},
    {"item": "banana", "price": 0.5, "quantity": 5},
    {"item": "orange", "price": 0.75, "quantity": 2}
]


print(calculate_total_cost(cart))  # prints the total cost of items in the cart, which is 6.25
```

```python
# # ckeck if a string is an palindrome  

# palindrome = "madam" --- same from fromt and back 

def is_palindrome(s):
    s =s.lower().replace(" ","")# remove spaces and convert to lowercase using (.lower())
    return s == s[::-1]  # check if the string is equal to its reverse [::-1 ]  is used to reverse what ever ewe hve 

print(is_palindrome("madam"))  # prints True
print(is_palindrome("hello"))  # prints False
```

```python
# calculate the factoraial of numbers using te recusion 
# 3 factorial = 5! = 5 * 4 * 3 * 2 * 1 = 120

def factorial(n):
    if n==0 :
        return 1 
    else :
        return n * factorial(n-1)  # recursive call to calculate factorial


print(factorial(5))  # prints 120
```

```python
# funvtion to read a file an d count the frequrnct of each
# here we iwll count how many times each word appears in the file
# 3 we will be using "open func" provide by outhin req to open any kinf of file 

def count(file_path):
    wordcount = {} # empty dict
    with open(file_path, 'r') as file:
        for line in file:
            words = line.split() # we are splitting the line into words based on the character ---using this line we will get the words 
            for word in words : # acces the words in the line
                word = word.lower().strip('.,!?;:\'') # we perfrom thetlower operation on an swords avilable-- "Strips()" remove the white spaces in the sentnecea
                wordcount[word]= wordcount.get(word,0)+1 # we use get method to get the value of the word in the dictionary, if it does not exist, it returns 0, then we add 1 to the count
    return wordcount  # return the dictionary containing word counts
#
# wordcount[word] acces the word

filepath = 'sample.txt'  # replace with your file path
wordfreq =  count(filepath)  # calling the count function to get word frequency
print(wordfreq)  # prints the word frequency dictionary
```

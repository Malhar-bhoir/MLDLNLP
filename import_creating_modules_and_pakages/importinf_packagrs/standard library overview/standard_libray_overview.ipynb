```python
## data seralizaion -- use for converting the file from json to python dict and vice versa
# this is useful when we get some response from the api in json format and we want to convert it into python dict
import json
data = {"ron": "name", "age": 20}

# convert this dictionary to json
json_str = json.dumps(data)  # use for conversion
print(json_str)
print(type(json_str))
# convert json again to dict

parsed_data = json.loads(json_str)
print(parsed_data)
print(type(parsed_data))
 
```

```python
# handling the missing values
df.isnull().any(axis=1 ) # here axis = 1 indicate for every record it willl show hweer ther is the missing value and wherethere is not 
```

```python
df.isnull().sum()  # here it will show the total number of missing values in each column
```

```python
df_filled = df.fillna(0) # fillna indicates to fill the missing values with the given value here 0
df
```

```python
# filling the missing values with the measn of the column
# 3 HERE HWERE EVER the NAN / Null values are present we will replacethat with the "measn" of the sales 
df['Salesfillna'] = df['Sales'].fillna(df['Sales'].mean())
df
```

```python
# How to vhane an datatype
df.dtypes
```

```python
# suppose wewant t ochange the floar64 of the sales into the number or we  can say as an "int"


# --1--- # rennaimg the column 

df = df.rename(columns = {'Date ' : 'Sales Date'})
df.head()
```

```python
df['value_new']  = df['Value'].fillna(df["Value"].mean()).astype(int)
df.head() # -- here inside ---astype() ---we providethe parameter as teh the goal the value type tha we that we want 
```

```python
# aplllying the operations to any of the functions 
df['NewVlaue']  = df['Value'].apply(lambda x:x*2)
```

```python
df
```

```python
# data aggreagration and grouping ---suppose we want to know the total sales per category
df.groupby('Category')['Sales'].sum()

# we can also perform mean based on thr proeducts 
grouped_mean = df.groupby('Product')['Value'].mean()
# her product is the categorical value (means it has diff categorieas as product 1 pro 2etx)
# we do also need to find the mean to determine which category has the mosre sales 
```

```python
print(grouped_mean)
```

```python
 # performing the grouped based on the two categorical product 
Gsum  = df.groupby(["Product" , "Region"])['Value'].sum()
print(Gsum)
```

```python
# applying the multiple agrregreating Function 
Group_Agregration  = df.groupby('Product')['Value'].agg(['mean' , 'sum'])
Group_Agregration
```

```python
# merging and joing the dataframe  
df1 = pd.DataFrame( { 'key' : ['a' , 'b ' , 'c'], 'value 1 ' : [10 , 20 , 30 ] } )
df2 = pd.DataFrame( { 'key' : ['a' , 'b ' , 'c','d'], 'value 2 ' : [100 , 200 , 300,400 ] } )

df1
```

```python
# mergin the datafram based on th e key coluumns 
# the parameter to the merge are ----two fraes that we want to merge 
# ---- 2---- then the next we can take os " on" indicate which columst obe merged 
#--3-- tehn the 3rd indicateteh joint type  ---how = inner--

pd.merge(df1 , df2 , on = "key" , how = 'inner') # on which column we want to work and which way por which operation to perform
```

```python
pd.merge(df1 , df2 , on = "key" , how = 'outer') # on which column we want to work and which way por which operation to perform
```

```python
#  attritibutes of Array
arr = np.array([[1, 2, 3], [4, 5, 6]])

print("Array:\n", arr)
print("Shape:", arr.shape) # Output: (2, 3)
print("Number of dimensions:", arr.ndim) # Output: 2
print("Size (number of elements):", arr.size) # Output: 6
print("Data type:", arr.dtype) # Output: int64.(may.vary.based.on.platform)
print("Item size (in bytes):", arr.itemsize) # Output: 8 (may vary based on os
```

```python
# numpy vectorize operation 
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.array([10, 20, 30, 40, 50])

# addition  operation 
print("addition:" , arr1 + arr2)

# sub 
print("sub : " , arr1 - arr2)

# div 

print("div:", arr1 / arr2) 
 # in mumpy every numpy operation provide the same class 
```

```python
# univeral Function , func tha applies to the entire function 
# these are performed as ----np.fun------\\\


arr = np.array([1,2,3,4,5])

#square root 
print(np.sqrt(arr))

# ecponwntinal 

print(np.exp(arr))

# sine 

print(np.sin(arr))


# log 

print(np.log(arr))
```

```python
# stastical concept -- normalization 
# to have mean 0 and standard deviation 1

data = np.array([1,2,3,4,5])


# calculation 
mean = np.mean(data)
std = np.std(data)

# 
# normalaization 

nor  = (data - mean) / (std)
print('nor' , nor)
```

### Data Visualization With Matplotlib

Matplotlib is a powerful plotting library for Python that enables the creation of static, animated, and interactive
visualizations. It is widely used for data visualization in data science and analytics. In this lesson, we will cover
the basics of Matplotlib, including creating various types of plots and customizing them.

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]

plt.plot(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title('titel')
```

```python
x= ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
y = [10, 20, 25, 30, 40]

plt.plot(x,y , color = 'red' , linestyle = '--' , marker  = 'o' , linewidth = 2  , markersize = 9 ) # marker is used to show the point on the graph
plt.xlabel("Days")  
plt.ylabel("Sales")
```

```python
# mulltiple plots 

x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [2,3,5,7,11]
# 3 'figure_size' indicate the size of the chart 
# subplot(1,2,1) indicate --1 --- row , 2--- column , the 3rd ---element -- in this case is the --1-- indiacte in which plcae we will be plottingthat so 
# so here {-,-, 1} indiacate we will plot that in the first space 

plt.figure(figsize = (9,6))# sixe pf the plot 

# 1
plt.subplot(1,2,1)
plt.plot(x,y1, color = 'red' , marker = 'o' , linestyle = '--')
plt.title('First Plot')

# 2 
plt.subplot(1,2,2)
plt.plot(x ,y2  , color = 'blue')
plt.title('secomd')

```

✅ 1. What is an Iterator?

An iterator is something you can loop through one item at a time.

Simple meaning:

Iterator = a machine that gives you values one by one when you ask for them.

How to get an iterator?

Use iter()

Get next value using next()

```python
numbers = [1, 2, 3, 4, 5]

it = iter(numbers)

print(next(it))
```

If you call next(it) again → error (no more items).

Where do you use iterators?

In loops (for loop automatically uses iterator)

Working with streams

Large datasets

✅ 2. What is a Generator?

A generator is a special function that remembers where it left off and gives values one-by-one using yield.

Simple meaning:

Generator = a lazy function that produces values only when needed.

Why use?

Saves memory

Good for big data

```python
## eample of te yeild function 
def generator_function():
        yield 1 
        yield 2
        yield 3
gen = generator_function()

print(next(gen))
```
